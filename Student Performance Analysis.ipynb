##  1Ô∏è‚É£ How does weekly study time (StudyTimeWeekly) affect GPA?

```python
sns.regplot(data=SPA, x="StudyTimeWeekly", y="GPA", 
           marker="x",color=".3",line_kws=dict(color="r"),scatter_kws={"alpha": 0.4});
plt.title("Relationship between Study Time Weekly and GPA")
plt.xlabel("Study Time Weekly (Hours)")
plt.ylabel("GPA")
plt.show()
```

### ‚úÖ Result: How does weekly study time affect GPA?

The scatter plot with a regression line shows a **slight positive relationship** between weekly study time (`StudyTimeWeekly`) and GPA.

- **Observation:** Students who spend more hours studying weekly tend to have a slightly higher GPA on average.
- **However**, the data points are widely scattered around the trend line, indicating that study time alone does not strongly determine GPA.
- **Recommendation:** Other factors such as absences, parental support, or extracurricular activities may also play significant roles and should be analyzed further.

_Source: Dataset from Kaggle ‚Äî Students Performance_

## 2Ô∏è‚É£ What is the relationship between absences (Absences) and academic performance (GPA)?

```python
sns.regplot(data=SPA, x="Absences", y="GPA", 
           marker="+",color='b',line_kws=dict(color='r'),
            scatter_kws={"alpha": 0.4});
plt.title("Relationship between Absences and GPA")
plt.xlabel("Absences")
plt.ylabel("GPA")
plt.show()
```

### ‚úÖ Result: What is the relationship between absences and GPA?

The scatter plot with a regression line shows a **clear negative relationship** between the number of absences (`Absences`) and GPA.

- **Observation:** Students with more absences tend to have a lower GPA.
- The downward trend in the regression line indicates that higher absence rates are strongly associated with lower academic performance.
- This suggests that regular attendance may have a significant positive impact on students‚Äô academic outcomes.

_Source: Dataset from Kaggle ‚Äî Students Performance_

## 3Ô∏è‚É£ Is participation in extracurricular activities (Sports, Music, Volunteering) associated with higher or lower GPA?

```python
sns.boxplot(data=SPA,x='Sports',y='GPA')
plt.title("GPA by Sports Participation")
plt.show()

sns.boxplot(data=SPA,x='Music',y='GPA')
plt.title("GPA by Music Participation")
plt.show()

sns.boxplot(data=SPA,x='Volunteering',y='GPA')
plt.title("GPA by Volunteering Participation")
plt.show()
```

### ‚úÖ Result: Is participation in extracurricular activities associated with higher or lower GPA?

The boxplots show the GPA distributions for students who participate vs. do not participate in **sports**, **music**, and **volunteering** activities.

- **Sports:** Students who participate in sports show a slightly higher median GPA than those who do not.
- **Music:** Participation in music activities appears to be associated with a slightly higher GPA compared to non-participants.
- **Volunteering:** Volunteering shows a small positive difference in GPA compared to students who do not volunteer.

However, the overlap between the groups suggests that while there is a positive trend, participation alone may not be a strong predictor of GPA. Other factors likely contribute to overall academic performance.


## 4Ô∏è‚É£ How does academic performance (GPA) compare across different ethnicities (Ethnicity)?

```python
sns.boxplot(data=SPA,x='Ethnicity',y='GPA')
plt.title("GPA by Ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("GPA")
plt.show()
```

### ‚úÖ Result: How does academic performance (GPA) compare across different ethnicities?

The boxplot shows the distribution of GPA for students across different ethnicities (`Ethnicity`).

- **Observation:** The median GPA appears to be relatively similar across the different ethnicity groups (0, 1, 2, 3).
- There are slight differences in the spread and quartiles, but overall, the GPA distribution is fairly consistent among the groups.
- This suggests that ethnicity in this dataset does not show a strong association with significant GPA differences by itself.

Further analysis with other factors (like parental education or study time) may help reveal more detailed insights.
